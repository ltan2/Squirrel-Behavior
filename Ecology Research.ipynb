```python
import pandas as pd
import numpy as np
```

```python
squirrel_df = pd.read_csv("ecologylabjupyter.csv")

#cleaned the data first
squirrel_df = squirrel_df.drop(columns=['Proximity to safety','Proximity to nearest edge','Proximity to human structures'])
squirrel_df = squirrel_df.drop(columns=['Clouds','University','Biome','Wind','Precipitation'])

#print(squirrel_df)
```

```python
split_df= squirrel_df['Date'].str.split(',').apply(pd.Series)
split_df.columns = ['Day','Date','Year']
#print(split_df)

squirrel_df = squirrel_df.drop(columns=['Date'])
squirrel_df = squirrel_df.join(split_df, how="outer")
#print(squirrel_df)
```

```python
tmp_dfs = squirrel_df

#split by time
time_df = tmp_dfs['Time'].str.split(' ').apply(pd.Series)
time_df.columns = ['Time val','AM/PM']

tmp_dfs = tmp_dfs.drop(columns=['Time'])
tmp_dfs = tmp_dfs.join(time_df,how="outer")
tmp_dfs = tmp_dfs.groupby(tmp_dfs['AM/PM'])

#for name, tmp_df in tmp_dfs:
   # print(name)
    #print(tmp_df)
```

```python
AM_df = tmp_dfs.get_group("AM")
#print(AM_df)

specificTime_df = AM_df['Time val'].str.split(':').apply(pd.Series)
specificTime_df.columns = ['Hour','Minute','Seconds']
#print(specificTime_df)
```

```python
AM_df = AM_df.drop(columns=['Time val'])
AM_df = AM_df.join(specificTime_df,how="outer")
#print(AM_df)
```

```python
array = np.zeros(12) # representing time from 12am -11am
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
days_num = np.zeros(7)
time = []
day_name = []
numRows = []
numForaging = []
numVigilance = []
numAlertFeeding = []
numSocial = []
numOthers = []

AMhour_dfs = AM_df.groupby(AM_df['Hour'])

for i in range(1,13):
    if str(i) in AMhour_dfs.groups.keys():
        AMhour_df = AMhour_dfs.get_group(str(i))
        array[i-1] = len(AMhour_df)
        days_dfs = AMhour_df.groupby(AMhour_df['Day'])
        for name,df in days_dfs:
            index = days.index(name)
            time.append(str(i) + " am")
            day_name.append(name)
            numRows.append(len(df))
            numForaging.append((df['Total foraging'].sum())/ len(df))
            numVigilance.append((df['Total vigilance'].sum()) / len(df))
            numAlertFeeding.append((df['Total Alert Feeding'].sum()) / len(df))
            numSocial.append((df['Total social'].sum()) / len(df))
            numOthers.append((df['Total other'].sum()) / len(df))
            

numVigilance = pd.DataFrame(numVigilance)
numForaging = pd.DataFrame(numForaging)
numAlertFeeding = pd.DataFrame(numAlertFeeding)
numSocial = pd.DataFrame(numSocial)
numOthers = pd.DataFrame(numOthers)

time = pd.DataFrame(time)
day_name = pd.DataFrame(day_name)
numRows = pd.DataFrame(numRows)
new_df = pd.concat([time,day_name,numRows,numForaging,numVigilance,numAlertFeeding,numSocial,numOthers],axis=1)
new_df.columns = ['Time','Day','Number of squirrels','Average Foraging','Average Vigilance','Average Alert Feeding','Average social','Average others']

print(new_df)

```
